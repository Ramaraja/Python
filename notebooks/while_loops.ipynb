```python
# break - to stop the loop 
x = 2
while x < 10:
    print ("x is less than 10")
    x+=1
    if x==5:
        break
```

```python
# continue - to skip the logic

x = 0
while x < 10:
    print ("counter-", x)
    x+=1
    if x==5:
        print ("skip")
        continue
```
