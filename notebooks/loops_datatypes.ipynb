```python
# For loops

l = [1,2,3,4,5,6,7,8,9]
```

```python
for item in l:
    print (item)
```

```python
s = "this is a test"
for char in s:
    print (char)
```

```python
# tuple

t = (1,2,3,4)
for item in t:
    print (item)
```

```python
# unpacking

li = [(1,2,3),(3,4),(5,6)]
```

```python
for (t1,t2) in li:
    print (t2)

    
```

```python
t1,t3 = (1,2)
print (t1)
```

```python
# dict

d = {"d": 1, "b":2, "c": 3}

for item in d:
    print (item)
```

```python
for k,v in d.items():
    print (k,v)
```

```python
# dictionary keys and values

for k in d.keys():
    # print (k)
    print (d[k])
```

```python
# values

for v in d.values():
    print (v)
```
